```python
j = 0
# 1 Living
# 0 Non-Living
for j in range(1000):
    df.at[j,'Object'] = objects[random.randint(0,len(objects)-1)]
    if(living_objects.count(df.at[j,'Object'])):
        df.at[j,'LNL'] = 1
    else:
        df.at[j,'LNL'] = 0
```

```python
for j in range(1000):
    df.at[j,'Height'] = random.randint(10,200)
    df.at[j,'Distance'] = random.randint(50,1000)
    df.at[j,'Yolo_Decision'] = random.randint(0,1)
    df.at[j,'Speech_Decision'] = random.randint(0,1)
    df.at[j,'Final_Decision'] = random.randint(0,1)
#1 Stop the Car
#0 Don't stop the Car
# 1 Living
# 0 Non- Living
df
```

```python
for j in range(1000):
    #Non-Living Object smaller than ground clearance
    if(df.at[j,'Height']<25 and df.at[j,'LNL']==0):
        df.at[j,'Yolo_Decision'] = 0
    
    elif(df.at[j,'Height']>=25 and (df.at[j,'LNL']==0 or df.at[j,'LNL']==1) and df.at[j,'Distance']>500 ):
        df.at[j,'Yolo_Decision'] = 0

    #Object Detected by User but not by Yolo
    elif((df.at[j,'Height']>=200) and (df.at[j,'LNL']==0 or df.at[j,'LNL']==1) and df.at[j,'Distance']<500 ):
        #if(random.randint())
        df.at[j,'Yolo_Decision'] = 0 #Yolo Don'tStop
        df.at[j,'Speech_Decision'] = 1 #User Says Stop 
    
    #Object Detected by User but not by Yolo
    elif((df.at[j,'Height']>=25 and df.at[j,'Height']<=200) and (df.at[j,'LNL']==0 or df.at[j,'LNL']==1) and df.at[j,'Distance']<500 ):
        #if(random.randint())
        df.at[j,'Yolo_Decision'] = 1 #Yolo Stop

    #Here object is visible to YOLO But not to User
    elif(df.at[j,'Height'] <25 and df.at[j,'LNL']==1 and df.at[j,'Distance']<500):
        df.at[j,'Yolo_Decision'] = 1
        if(df.at[j,'Distance']<200):
            df.at[j,'Speech_Decision'] = 0
```

```python
# Importing SVC classifier
from sklearn.svm import SVC

# For Accuracy Calculation
from sklearn.metrics import accuracy_score


from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
X_train, X_test, y_train, y_test = train_test_split(df.drop(['Final_Decision','Object'],axis=1), df['Final_Decision'], test_size=0.30,random_state=101)

#Without Scaling the Features Calculating Accuracy of SVM  

#To Display the Graph for Accuracy Comparison
accList = []
model = ["SVM Feature Not Scaled","SVM Feature Scaled","Logistic Regression","Decision Tree (Entropy)","Decision Tree (Gini Index)"]


l = ['linear', 'poly', 'rbf', 'sigmoid'] #List of Kernel Values 
C = [1,10] #Values of C parameter


maxK = None #Kernel Value for max Accuracy
maxC = None #C Value for max Accuracy
maxM = 0 #Max Accuracy Value for Testing Dataset
    # instantiate classifier with default hyperparameters

print("\n**** SVM without Features Scaling ****\n")
for i in l:
    for c in C:

        svc=SVC(kernel=i,C=c) 
       
        #Training the Model
        svc.fit(X_train,y_train)


        # Prediction using Trained Model
        y_pred=svc.predict(X_test)


        # Accuracy Score of SVM with Parameter
        print('Model accuracy score with default kernel = ', i,' and C = ',c,' hyperparameters: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
        # print the scores on training and test set
        if(accuracy_score(y_test, y_pred)>maxM):
            maxM = accuracy_score(y_test, y_pred)
            maxK = i
            maxC = c

        print('Training set score: {:.4f}'.format(svc.score(X_train, y_train)))

        print('Test set score: {:.4f}'.format(svc.score(X_test, y_test)))

        print("The Precision Score is : ",precision_score(y_test, y_pred))

        print("The F1 Score : ",f1_score(y_test, y_pred))

        print("\n")
    print("************************************************************************************")

print("\nFor SVM using unScaled Features Maximum Accuracy is ",maxM)
print("\nFor Kernel ",maxK)
print("\nFor C = ",maxC)
accList.append(maxM)
```
